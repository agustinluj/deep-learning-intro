```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Importar y comprobar las 9 primeras imágenes de perros

# Ruta al directorio que contiene sus datos de imagen
data_dir = r"C:/Users/Agustín/Desktop/4Geeks/Clases/26. Intro to Deep Learning/train/train"

# Obtener una lista de todos los nombres de archivos de imágenes de perros
file_list = os.listdir(data_dir)
cat_list = [i for i in file_list if 'cat' in i]
dog_list = [i for i in file_list if 'dog' in i]

```

```python
dog_images = []
for i in range(9):
    img_path = os.path.join(data_dir, dog_list[i])
    img = image.load_img(img_path)
    img_array = image.img_to_array(img)
    img_array /= 255.0
    dog_images.append(img_array)
```

```python
cat_images = []
for i in range(9):
    img_path = os.path.join(data_dir, cat_list[i])
    img = image.load_img(img_path)
    img_array = image.img_to_array(img)
    img_array /= 255.0
    cat_images.append(img_array)
```

```python
plt.figure(figsize=(12, 8))

for index, image in enumerate(dog_images[:3]):  # `image` holds each element in `dog_list`
    plt.subplot(3, 3, index + 1)
    plt.imshow(image)  # Use `image` directly
    plt.title('Dog Image')
    plt.axis('off')

for index, image in enumerate(cat_images[:3]):  
    plt.subplot(3, 3, index + 4)
    plt.imshow(image)  # Use `image` directly
    plt.title('Cat Image')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
